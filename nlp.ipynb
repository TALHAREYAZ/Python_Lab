```python
# Importing Transformer Model

from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

***This line imports the necessary modules from the transformers library. 'MBartForConditionalGeneration' is the model class for the MBART model, and 'MBart50TokenizerFast' is the tokenizer class.***

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
```

***This line initializes an instance of the MBartForConditionalGeneration model by loading the pre-trained weights for the "facebook/mbart-large-50-one-to-many-mmt" model. This model supports translation from English to multiple languages.***


```python
# Available Languages to Translate

# Arabic (ar_AR), Czech (cs_CZ), German (de_DE), English (en_XX), Spanish (es_XX), 
# Estonian (et_EE), Finnish (fi_FI), French (fr_XX), Gujarati (gu_IN), Hindi (hi_IN), 
# Italian (it_IT), Japanese (ja_XX), Kazakh (kk_KZ), Korean (ko_KR), Lithuanian (lt_LT), 
# Latvian (lv_LV), Burmese (my_MM), Nepali (ne_NP), Dutch (nl_XX), Romanian (ro_RO),
# Russian (ru_RU),Sinhala (si_LK), Turkish (tr_TR), Vietnamese (vi_VN), Chinese (zh_CN),
# Afrikaans (af_ZA),Azerbaijani (az_AZ),Bengali (bn_IN), Persian (fa_IR), Hebrew (he_IL),
# Croatian (hr_HR), Indonesian (id_ID), Georgian (ka_GE),Khmer (km_KH), Macedonian (mk_MK), Malayalam (ml_IN),
# Mongolian (mn_MN), Marathi (mr_IN), Polish (pl_PL),Pashto (ps_AF), Portuguese (pt_XX), Swedish (sv_SE),
# Swahili (sw_KE), Tamil (ta_IN), Telugu (te_IN), Thai (th_TH),Tagalog (tl_XX), Ukrainian (uk_UA), Urdu (ur_PK),
# Xhosa (xh_ZA), Galician (gl_ES), Slovene (sl_SI)
```

```python
#defining model and source language

tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt",src_lang="en_XX")
```

***Here, the MBart50TokenizerFast is initialized with the "facebook/mbart-large-50-one-to-many-mmt" pre-trained model and the source language set to English (src_lang="en_XX"). The tokenizer is responsible for tokenizing the input text.***

```python
# set input text
text="To work in a competitive environment and to give excellent performance in whatever jobs assigned. I want to work with a dynamic company where I can excel my skills and contribute to my company"
```

\***This line defines the input text that you want to translate. ***

```python
model_inputs = tokenizer(text, return_tensors="pt")
```

***Here, the tokenizer is used to tokenize the input text. The tokenizer converts the text into numerical inputs that the model can process. return_tensors="pt" specifies that the tokenizer should return PyTorch tensors as outputs.***

```python
# Translate from English to Malayalam
generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
)
```

***In this line, the model.generate() method is called to generate the translated tokens. The model_inputs syntax is used to pass the tokenized inputs to the model. forced_bos_token_id is set to the language code for Hindi (hi_IN), indicating that the model should generate translations in Hindi.***

```python
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

***This line decodes the generated tokens back into human-readable text. batch_decode() is used because the generate() method returns results in batches. The skip_special_tokens=True argument removes any special tokens from the decoded text.***

***The output of this code will be the translated sentence in Hindi based on the input English sentence provided.***
